Problem 1: Smart Light Controller


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_status(self, action):
        self.status = action.upper()
        print(f"{self.name} is {self.status}")

# Usage
light = SmartLight("Bedroom Light")
light.set_status("ON")
```

Problem 2: Employee ID Card System


```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Usage
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

Problem 3: Mobile Contact Record


```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Usage
c = Contact("Anita", "9876543210")
c.display_contact()
```

Problem 4: Product Price Tag Generator


```python
class PriceTag:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

# Usage
tag = PriceTag("Headphones", 2499)
tag.print_tag()
```

Problem 5: Movie Rating Display System


```python

class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def display_rating(self):
        print(f"Movie Name: {self.title}")
        print(f"Rating: {self.rating}")

# Usage
m = Movie("Inception", 4.8)
m.display_rating()
```

Problem 6: Delivery Address Manager


```python
class DeliveryManager:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

# Usage
delivery = DeliveryManager("Suman", "Hyderabad")
delivery.display_delivery()
```
