 Problem 1: Unique Words in a Sentence:


```python

# --- Input ---
sentence = "python is easy and python is powerful"

# --- Logic ---
words_list = sentence.split()      # Step 1: Break into individual words
unique_words = set(words_list)     # Step 2: Convert to set to remove duplicates

# --- Output ---
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

Problem 2: Highest Salary from Employee Data:


```python
# --- Input ---
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
}

# --- Logic ---
# Professional approach: using max() with a key argument
best_employee = max(employees, key=employees.get)
highest_salary = employees[best_employee]

# --- Output ---
print(f"Highest Salary: {best_employee} - {highest_salary}")
```

Problem 3: Find Maximum and Minimum Values:

```python
# --- Input ---
numbers = [45, 22, 89, 10, 66]

# --- Logic ---
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

# --- Output ---
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

Problem 4: Count Products Above a Price Threshold:


```python
# --- Input ---
prices = [450, 1200, 899, 1500, 300]

# --- Logic ---
count_above_1000 = 0
for price in prices:
    if price > 1000:
        count_above_1000 += 1

# --- Output ---
print(f"Products above 1000: {count_above_1000}")
```

Problem 5: Calculate Attendance Percentage:


```python
# --- Input ---
attendance = ["P", "P", "A", "P", "P"]

# --- Logic ---
total_days = len(attendance)
days_present = attendance.count("P")
percentage = (days_present / total_days) * 100

# --- Output ---
print(f"Attendance Percentage: {percentage}%")
```

Problem 7: Count Character Frequency:


```python
# --- Input ---
text_input = "python" # Expanding to 'python' as per example logic, or a string with repeats

# Let's use a string with repeats to demonstrate frequency better
target_string = "ppython"

# --- Logic ---
frequency_map = {}
for char in target_string:
    if char in frequency_map:
        frequency_map[char] += 1
    else:
        frequency_map[char] = 1

# --- Output ---
print(frequency_map)
```

Problem 10: Calculate Average Marks:


```python
# --- Input ---
marks = [70, 80, 60, 90, 62.5] # Example data to match output 72.5

# --- Logic ---
total_marks = sum(marks)
count = len(marks)
average = total_marks / count

# --- Output ---
print(f"Average Marks: {average}")
```
