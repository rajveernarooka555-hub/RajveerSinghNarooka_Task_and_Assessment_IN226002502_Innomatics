Problem Statement 1: Smart Parking Lot Management System


```python
def manage_parking(capacity, logs):
    current_count = 0
    for log in logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT":
            current_count -= 1

    status = "Full" if current_count >= capacity else "Available"

    print(f"Currently Parked Vehicles: {current_count}")
    print(f"Parking Status: {status}")

# Test
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

 Problem Statement 2: Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):
    # Base time: 5 mins per km
    eta = distance * 5

    # Traffic Adjustments
    traffic_delays = {"Low": 0, "Medium": 10, "High": 5}
    eta += traffic_delays.get(traffic, 0)

    # Weather Adjustments
    weather_delays = {"Clear": 0, "Rainy": 10, "Snowy": 20}
    eta += weather_delays.get(weather, 0)

    print(f"Estimated Delivery Time: {eta} minutes")

# Test
estimate_delivery_time(8, "High", "Rainy")
```

 Problem Statement 3: Movie Theatre Seat Occupancy

```python
def analyze_occupancy(total_seats, booked_seats):
    count = len(booked_seats)
    percentage = (count / total_seats) * 100

    if percentage == 100:
        status = "Housefull"
    elif percentage >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print(f"Occupancy: {percentage}%")
    print(f"Show Status: {status}")

# Test (150 entries in list)
analyze_occupancy(200, [1]*150)
```

 Problem Statement 4: Cloud Server Load Classification System


```python
def classify_server_load(readings):
    avg_load = sum(readings) / len(readings)

    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg_load}%")
    print(f"Server Status: {status}")

# Test
classify_server_load([45, 60, 70, 85, 90])
```

 Problem Statement 5: Smart Classroom Resource Usage Monitor


```python
def monitor_resources(usage_dict):
    overused = [res for res, hours in usage_dict.items() if hours > 8]
    energy_alert = "Yes" if overused else "No"

    print(f"Overused Resources: {', '.join(overused) if overused else 'None'}")
    print(f"Energy Alert: {energy_alert}")

# Test
monitor_resources({"Projector": 6, "AC": 9, "Lights": 4})
```

 Problem Statement 6: Online Event Registration Capacity Controller


```python
def manage_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"

    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# Test
manage_registration(100, 105)
```
